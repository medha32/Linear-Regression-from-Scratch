```python
import numpy as np
import pandas as pd
```

```python
path = '/content/Housing.csv'
df = pd.read_csv(path)
df
```

```python
X = df.iloc[:, 0].values
Y = df.iloc[:, 1].values
```

```python
print(X.shape)
print(Y.shape)
```

```python
import numpy as np
X_raw = df[['area', 'bedrooms']].values
```

```python
def compute_cost(X, Y, weights):
    n_samples = len(Y)
    predictions = X.dot(weights)
    cost = (1 / (2 * n_samples)) * np.sum((predictions - Y) ** 2)

    return cost
```

```python
def gradient_descent(X, Y, weights, learning_rate, n_iterations):
    n_samples = len(Y)
    cost_history = []

    for i in range(n_iterations):
        predictions = X.dot(weights)
        errors = predictions - Y
        gradients = (1 / n_samples) * X.T.dot(errors)
        weights -= learning_rate * gradients
        cost = compute_cost(X, Y, weights)
        cost_history.append(cost)
    return weights, cost_history
```

```python
def linear_regression(X, Y, learning_rate=0.01, n_iterations=1000):
    X_b = np.c_[np.ones((X.shape[0], 1)), X]
    weights = np.zeros(X_b.shape[1])
    optimized_weights, cost_history = gradient_descent(X_b, Y, weights, learning_rate, n_iterations)
    return optimized_weights, cost_history
```

```python
learning_rate = 0.01
n_iterations = 1000
weights, cost_history = linear_regression(X_raw, Y, learning_rate, n_iterations)
weights, cost_history[-1]
```

```python
X_normalized = (X_raw - X_raw.mean(axis=0)) / X_raw.std(axis=0)
weights, cost_history = linear_regression(X_normalized, Y, learning_rate, n_iterations)
weights, cost_history[-1]
```
